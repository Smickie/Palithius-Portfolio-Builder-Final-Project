```python
def get_br_pe(x):

    filt = instruments.loc[instruments["name"] == x]["branchId"]

    temp = instruments.loc[instruments["branchId"] == filt[filt.index[0]]]

    pes = []

    for i in temp["name"]:
        pe = get_pe(i)
        if pe >= 0:
            pes.append(pe)

    avg = sum(pes) / len(pes)
    mini = min(pes)
    maxi = max(pes)
    
    return [avg, mini, maxi]
```
